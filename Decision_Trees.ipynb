# **What is a Decision Tree ?**

**Decision tree** is a type of **Supervised learning algorithm** that is mostly used in classification problems. It works for both categorical and continuous input and output variables.

# **How to split?**

The Decision of making strategic splits heavily affects a tree's accuracy. The decision criteria is different for classification and regresssion trees.

Decision trees use multiple algorithms to decide to split a node in two or more sub-nodes. The creation of sub-nodes increases the homogeneity of resultant sub-nodes. In other words, we can say that purity of the node increases with respect to the target variable. Decision tree splits the nodes on all available variables and then selects the split which results in most homogenous sub-nodes.

#### **Methods to determine best split:**
* Information Gain
* Gini
* Chi Square
* Reduction in variance

# **Implementing a Simple Decision Tree:**
A Decision Tree is a type of machine learning algorithm that is used for both classification and regression tasks. It's like a flowchart that makes decisions based on the features of the data. Each node in the tree represents a test on an attribute (e.g., whether a fruit is red), each branch represents the outcome of the test, and each leaf node represents a class label (decision taken after computing all attributes).

Let's consider a simple example using the Decision Tree Classifier in Python. We'll use a small dataset of fruits characterized by their color and size to determine whether they are apples or bananas.

**Step 1: Install and Import Necessary Libraries** You need scikit-learn for the Decision Tree algorithm and pandas for handling the data.

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

**Step 2: Create a Dataset** Let's upload a small dataset of fruits.

**Dataset Link:** https://drive.google.com/file/d/1pshM0WJ3jPVnWQdzHTlRXQVu4OaZVWJz/view?usp=sharing

```python
df = pd.read_csv("/content/Fruit Classification.csv")
df
```

**Step 3: Preprocess the Data** We need to convert categorical data into numerical form since decision trees in scikit-learn do not handle categorical variables directly.

```python
df = df.drop('Unnamed: 0', axis = 1)
df
```

```python
df['Color'] = df['Color'].map({'Green': 0, 'Yellow': 1, 'Red': 2})
df['Size'] = df['Size'].map({'Small': 0, 'Medium': 1, 'Large': 2})
df['Fruit'] = df['Fruit'].map({'Apple': 0, 'Banana': 1})
```

**Step 4: Create and Train the Decision Tree Model** Split the data into train and test sets and then fit the model.

```python
X = df[['Color', 'Size']] # Features
y = df[['Fruit']] # Target Variable

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

**Step 5: Make Predictions** Now, let's use our model to predict the fruit type based on color and size.

```python
# Prdicing a fruit: Red and Small
prediction = model.predict([[2,0]])
prediction = 'Apple' if prediction == 0 else 'Banana'
print(prediction)
```

```python
test_predictions = model.predict(X_test)
```

**Step 6: Evaluating the Model** Let's evaluate our model's accuracy.

```python
# Measure Accuracy of the Classifier
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, test_predictions) * 100)
```

This is more of an illustrative example and not a good practice for actual model evaluation. We should typically use a separate train test dataset otherwise there will be a high risk of overfitting.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt

# plot The Decision Tree
plt.figure(figsize =(10,8))
plot_tree(model, filled=True, feature_names=['Color','Size'], class_names =['Apple','Banana'])
plt.show()
```

This simple example demonstrates how a decision tree is used for classification. In practice, decision trees can handle much more complex datasets and are the basis for more advanced algorithms like Random Forests.
